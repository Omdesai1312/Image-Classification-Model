```python
from icrawler.builtin import GoogleImageCrawler

players = ["Ma Long", "Fan Zhendong", "Timo Boll", "Truls Moregard"]

for player in players:
    google_crawler = GoogleImageCrawler(storage={"root_dir": f"dataset/{player.replace(' ', '_')}"})
    google_crawler.crawl(keyword=f"{player}", max_num=300)
```
